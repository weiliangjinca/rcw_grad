## testing convolutions
We will have a convolution matrix for a circle and then convolve it with a uniform real space field

```python
import numpy as np
import matplotlib.pyplot as plt
from convmat2D import *
#generate a picture array with a circle
Nx = 2*256
Ny = 2*256;

A = 1*np.ones((Nx,Ny));
ci = Nx/2-1; cj= Ny/2-1;
cr = np.round(0.35*Nx);
I,J=np.meshgrid(np.arange(A.shape[0]),np.arange(A.shape[1]));

dist = np.sqrt((I-ci)**2 + (J-cj)**2);
A[np.where(dist<cr)] = 12;

plt.imshow(A);
plt.show()
plt.colorbar

##fft
P = 20; Q = 20;
C1 = convmat2D(A, P, Q);
C2 = convmat2D_o(A, 2*P+1, 2*Q+1);
print(np.linalg.norm(C1-C2)) #make sure the two ways of writing convmat are the same...
print(C1.shape)
np.set_printoptions(precision=2)
print(C1)
plt.imshow(np.abs(C1));
plt.show()


## test 2d convmat on a homogeneous medium, should return a scaled identity matrix
eps_grid = np.ones((Nx,Ny));
C_uniform = convmat2D(eps_grid,3,3)
print(C_uniform)
```

```python
## convolve convolution matrix and recover real space result
convolving_field = np.ones((Nx,Ny))
#fourier decomp of convolving field
conv_uni = np.fft.fft(convolving_field)
print(np.max(conv_uni) )

cfieldfourier = np.zeros(((2*P+1)*(2*Q+1),))
cfieldfourier[(2*P+1)*(2*Q+1)//2+1] = 1;

cd = np.dot(C1, cfieldfourier)
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
plt.imshow(1681*np.abs(np.fft.ifft2(cd.reshape(2*P+1, 2*Q+1))))
plt.clim(1,12)
plt.colorbar();
plt.subplot(1,2,2)
plt.imshow(convolving_field*A)
plt.clim(1,12)
plt.colorbar();
```
